```python
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
import pandas

df = pandas.read_csv("business-price-indexes-december-2019-quarter-csv.csv", parse_dates=["Period"])

# create figure obj
p = figure(width=500, height=250, x_axis_type="datetime", sizing_mode='stretch_both')

# parse x-axis and y-axis
p.line(df["Period"], df["Data_value"], color="Orange", alpha=0.5)

output_file("plot1.html")
output_notebook()
show(p)
```

```python
# multiple glipse in one plot
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

p = figure(plot_width=500, plot_height=400)

p.title.text = "Earthquake"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"

p.xaxis.minor_tick_line_color = "red"
p.yaxis.minor_tick_line_color = "blue"

p.xaxis.axis_label = "Time"
p.yaxis.axis_label = "Value"

p.line([1,2,3,4,5], [4,6,5,3,2], line_width=2, color="green", alpha=0.5)
p.circle([i*2 for i in [1,2,3,4,5,6]], [4,6,5,3,2,3], size=8, color="red", alpha=0.5)

output_notebook()
show(p)
```

```python
# Single patches
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

p = figure(plot_width=400, plot_height=400)

# add patch renders with alpha on line width
p.patch([1,2,3,4,5], [6,7,8,7,3], alpha=0.5, line_width=2)

output_notebook()
show(p)
```

```python
# multiple patches
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

p = figure(plot_width=400, plot_height=400, sizing_mode='stretch_both')

# add patch renders with alpha on line width
p.patches([[1,3,2], [3,4,6,6]], [[2,1,4], [4,7,8,5]], color=["firebrick", "navy"], alpha=[0.8, 0.3], line_width=2)

output_file("multiple-patch.html")
output_notebook()
show(p)
```

```python
# Quantrant plot
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

p = figure(plot_width=400, plot_height=400)

p.quad(top=[2,3,4], bottom=[1,2,3], left=[1,2,3], right=[1.2, 2.5, 3.7], color="#B3DE69")

output_notebook()
show(p)
```
