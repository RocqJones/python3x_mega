```python
from bokeh.plotting import figure, output_file, show
import pandas

# create a df (data frame) and populate cols and rows
df = pandas.DataFrame(columns=["X","Y"])
df["X"] = [1,2,3,4,5]
df["Y"] = [5,6,4,5,3]

# save scatter plot in a variable and parse df
p = figure(title="Temperature Observations", x_axis_label="Day of observations", y_axis_label="Temperature")
p.circle("X", "Y", size=15, source=df)
# You can either generate an html of the plot or show it here

# 1. for html output_file
output_file("Scatter_plotting.html")
show(p)
```

```python
# glyph methods in bokeh.plotting (e.g circle, square) are all implicitly scatter-type functions to begin with
from bokeh.plotting import figure, show
import pandas as pd

df = pd.DataFrame({"X" :[1,2,3,4,5,6,7],
                   "Y": [23,43,32,12,34,54,33]})

p = figure(x_axis_label="Tempetature", y_axis_label="Day", 
           title="Day Temperature measurement")
p.circle("X", "Y", size=15, source=df)

show(p)
```

```python
# To work with dataframes, just pass in the columns like df.X and df.Y to the x and y args
from bokeh.plotting import figure, show, output_file
import pandas as pd

df = pd.DataFrame(columns=["X","Y"])

df["X"] = [1,2,3,4,5,6,7]
df["Y"] = [23,43,32,12,34,54,33]

p = figure()
p.scatter(df.X, df.Y, marker="circle")

from bokeh.io import output_notebook
output_notebook()

show(p)  # or output to a file...
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

p = figure(plot_width=500, plot_height=400,)

p.title.text = "Earthquake"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"

p.circle([1,2,3,4,5], [4,6,5,3,2], size=[i*2 for i in [4,6,8,10,12]], color="green", alpha=0.5)

output_notebook()
show(p)
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

p = figure(plot_width=500, plot_height=400,)

p.title.text = "Earthquake"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"

p.xaxis.minor_tick_line_color = "red"
p.yaxis.minor_tick_line_color = "blue"

p.xaxis.axis_label = "Time"
p.yaxis.axis_label = "Value"

p.circle([1,2,3,4,5], [4,6,5,3,2], size=8, color="green", alpha=0.5)

output_notebook()
show(p)
```

```python
# let's remove the bokeh logo and tools 
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook

p = figure(plot_width=500, plot_height=400, tools='')

p.title.text = "Earthquake"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"

p.xaxis.minor_tick_line_color = "red"
p.yaxis.minor_tick_line_color = "blue"

p.xaxis.axis_label = "Time"
p.yaxis.axis_label = "Value"

p.circle([1,2,3,4,5], [4,6,5,3,2], size=8, color="green", alpha=0.5)

output_file("graph_with_no_logo_and_tools.html")
output_notebook()
show(p)
```
