```python
import requests
from bs4 import BeautifulSoup

url = "https://pythontips.com/2013/07/30/20-python-libraries-you-cant-live-without/"

r = requests.get(url)
c = r.content

soup = BeautifulSoup(c,"html.parser")
all = soup.find_all("div",{"class":"entry-content"})
all

# print(all[0].find_all("a")[0].text)
# print(all[0].find_all("a")[1].text)

for items in len(all):
    print(items.find_all("a")[0].text)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://pythontips.com/2013/07/30/20-python-libraries-you-cant-live-without/"

r = requests.get(url)
c = r.content

soup = BeautifulSoup(c,"html.parser")
ally = soup.find_all("div",{"class":"entry-content"})
ally

for items in ally:
    print(items.find_all("p")[20].text)
```
